```python
import numpy as np
from scipy.special import gamma
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def f(z, c):
    return (-np.log(z))**(c-1) / gamma(c)
```

```python
x = np.linspace(0.01,0.995,200)
c = 0.1 / np.log(2)
```

```python
mu = 2**(-c)
sigma = 3**(-c) - 4**(-c)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, f(x, c), label="bethe heitler")
plt.plot(x, norm.pdf(x, mu, sigma), label="gaussian")
plt.xlabel("$E_f / E_i$")
plt.legend()
```
