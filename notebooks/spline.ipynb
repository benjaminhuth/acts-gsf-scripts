```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
```

```python
x = [ 1, 2, 4, 5 ]

y = np.exp(x)
```

```python
plt.plot(x,y)
```

```python
xnew = np.linspace(min(x), max(x), 300) 

spl = make_interp_spline(x, y, k=3)
ynew = spl(xnew)

plt.plot(xnew, ynew)
```
